Now let's deal with those missing values we found earlier, all were in the numerical columns (The protein columns). So it's better to impute them with the mean of their respective classes unlike their column's mean like in most datasets. 

```python
mean_values = df.groupby(by = ['class']).mean().reset_index()
mean_values
```

```python
class_column = 'class'

def fix_null_values(row, col):
    value = row[col]
    if pd.isnull(value):
        return mean_values.loc[mean_values[class_column] == row[class_column], col], True
    return value, False
```

```python
for i, row in df.iterrows():
    for col in df.columns:
        row[col], fixed = fix_null_values(row, col)
    df.iloc[i] = row
```

**The Model Evaluation Strategy**<br>


```python
#A cross validation method (Repeated K fold) will be used. 
from sklearn.model_selection import RepeatedKFold, GridSearchCV

cv_method = RepeatedKFold(n_splits=5,n_repeats=3,random_state=9)
```

**Hyperparameter Tunning**

Tuning can be performed efficiently if pipelines are used. In order to incoperate pipelines in feature selection using RFI, the following custom function is used. The references will be noted.

```python
from sklearn.base import BaseEstimator, TransformerMixin
# custom function for RFI feature selection inside a pipeline
# here we use n_estimators=100
class RFIFeatureSelector(BaseEstimator, TransformerMixin):
    
    # class constructor 
    # make sure class attributes end with a "_"
    # per scikit-learn convention to avoid errors
    def __init__(self, n_features_=10):
        self.n_features_ = n_features_
        self.fs_indices_ = None

    # override the fit function
    def fit(self, X, y):
        from sklearn.ensemble import RandomForestClassifier
        from numpy import argsort
        model_rfi = RandomForestClassifier(n_estimators=100,random_state=4)
        model_rfi.fit(X, y)
        self.fs_indices_ = argsort(model_rfi.feature_importances_)[::-1][0:self.n_features_] 
        return self 
    
    # override the transform function
    def transform(self, X, y=None):
        return X[:, self.fs_indices_]
```

```python
from sklearn.pipeline import Pipeline
```

**Decision Tree Parameter Tuning** <br> 
A Pipeline with a grid search will be used here. 
 

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
pipe_DT= Pipeline([('rfi_fs', RFIFeatureSelector()),
                    ('dt', DecisionTreeClassifier(random_state=111))])

params_pipe_DT = {'rfi_fs__n_features_': [20, 30, Data.shape[1]],
                  'dt__criterion': ['gini','entropy'],
                  'dt__max_depth': [5,12,14],
                  'dt__min_samples_split': [2,3]}

gs_pipe_DT = GridSearchCV(estimator=pipe_DT, 
                          param_grid=params_pipe_DT, 
                          cv=cv_method,
                          refit=True,
                          n_jobs=-2,
                          scoring='accuracy',
                          verbose=1) 

gs_pipe_DT.fit(D_train, t_train);
```

```python
gs_pipe_DT.best_params_
```

```python
gs_pipe_DT.best_score_
```

Please Note that the parameters dt_max_depth, dt_min_sample_split were randomly tried for different values and finally kept the above mentioned values as they gave the highest accuracy.

```python
# custom function to format the search results as a Pandas data frame
def get_search_results(gs):

    def model_result(scores, params):
        scores = {'mean_score': np.mean(scores),
             'std_score': np.std(scores),
             'min_score': np.min(scores),
             'max_score': np.max(scores)}
        return pd.Series({**params,**scores})

    models = []
    scores = []

    for i in range(gs.n_splits_):
        key = f"split{i}_test_score"
        r = gs.cv_results_[key]        
        scores.append(r.reshape(-1,1))

    all_scores = np.hstack(scores)
    for p, s in zip(gs.cv_results_['params'], all_scores):
        models.append((model_result(s, p)))

    pipe_results = pd.concat(models, axis=1).T.sort_values(['mean_score'], ascending=False)

    columns_first = ['mean_score', 'std_score', 'max_score', 'min_score']
    columns = columns_first + [c for c in pipe_results.columns if c not in columns_first]

    return pipe_results[columns]
```

```python
results_DT = get_search_results(gs_pipe_DT)

results_DT_20_features = results_DT[results_DT['rfi_fs__n_features_'] == 20.0]

alt.Chart(results_DT_20_features, 
          title='DT Performance Comparison with 20 Features'
         ).mark_line(point=True).encode(
    alt.X('dt__min_samples_split', title='Min Samples for Split'),
    alt.Y('mean_score', title='Accuracy', scale=alt.Scale(zero=False)),
    alt.Color('dt__max_depth:N', title='Max Depth')
)
```

**K-Nearest Neighbor Parameter Tunning** <br> 
A pipeline with a Grid Search is used here with the cross validation method defined earlier. 

```python
from sklearn.neighbors import KNeighborsClassifier

pipe_KNN = Pipeline(steps=[('rfi_fs', RFIFeatureSelector()), 
                           ('knn', KNeighborsClassifier())])

params_pipe_KNN = {'rfi_fs__n_features_': [10, 20, Data.shape[1]],
                   'knn__n_neighbors': [1, 5, 10],
                   'knn__p': [1, 2]}

gs_pipe_KNN = GridSearchCV(estimator=pipe_KNN, 
                           param_grid=params_pipe_KNN, 
                           cv=cv_method,
                           refit=True,
                           n_jobs=-2,
                           scoring='accuracy',
                           verbose=1) 

gs_pipe_KNN.fit(D_train, t_train);
```

```python
gs_pipe_KNN.best_params_
```

Different number of neighbours, distance measure were tried for and the above combination yielded more accuracy.
